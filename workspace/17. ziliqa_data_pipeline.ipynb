## 2.1 ZILIQA Train Data - Processed for Flair

- Ï†ÑÏ≤òÎ¶¨ Ïù¥ÌõÑ ÍπÄÎèÑÏòÅ ÌïôÏö∞ÏóêÍ≤å Ï†ÑÎã¨ÌïòÏó¨ ÌïôÏäµÌï† Îç∞Ïù¥ÌÑ∞

```python
#tweet data 2022.04.14~2022.04.28
date="2022-04-"
d_concat = data_concat()
ziliqa_df = []

for i in range(14,29):
    ziliqa_df.append(pd.read_csv('./ziliqa/ZIL OR ziliqa_'+date+str(i), 
                                  encoding='utf-8', on_bad_lines='skip').drop('number',axis=1))

ziliqa_data = d_concat.concat(ziliqa_df)
```

```python
flair_ziliqa = ziliqa_data[['date','tweet']]
```

```python
sentimental_tweet = sentimental_analysis(flair_ziliqa)
```

```python
sentimental_tweet.process('flair')
flair_processed_ziliqa = sentimental_tweet.twitter_data.drop('vader',axis=1)
flair_processed_ziliqa
```

```python
flair_processed_ziliqa.to_csv('flair_processed_ziliqa_train_data')
```
