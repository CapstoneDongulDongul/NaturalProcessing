```python
#tweet data 2022.04.14~2022.04.28
date="2022-04-"
d_concat = data_concat()
results =[]
for i in range(14,29):
    data1 = pd.read_csv('./bitcoin/bitcoin_'+date+str(i), encoding='utf-8').drop('number',axis=1)
    data2 = pd.read_csv('./bitcoin/btc_'+date+str(i), encoding='utf-8').drop('number',axis=1)
    r_data = d_concat.concat([data1,data2])
    print(data1.shape, data2.shape, r_data.shape)
    results.append(r_data)
```

```python
for result in results:
    print(result.shape)
```

```python
new_data = results[0]
for i in range(1,len(results)):
    files = [new_data,results[i]]
    new_data = d_concat.concat(files)
```

```python
new_data.reset_index(drop=True)
```

```python
sentimental_tweet = sentimental_analysis(new_data)
sentimental_tweet.process('vader')
sentimental_tweet.sentimental_with_vader()
```
