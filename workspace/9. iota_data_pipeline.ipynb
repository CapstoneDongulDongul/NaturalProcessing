## 2.1 IOTA Train Data - Processed for Flair

- 전처리 이후 김도영 학우에게 전달하여 학습할 데이터
- 2022.04.13~2022.04.27

```python
#tweet data 2022.04.14~2022.04.28
date="2022-04-"
d_concat = data_concat()
iota_df = []

for i in range(14,29):
    iota_df.append(pd.read_csv('./iota/iota OR miota_'+date+str(i), 
                                  encoding='utf-8', on_bad_lines='skip').drop('number',axis=1))

iota_data = d_concat.concat(iota_df)
```

```python
iota_data.isna().sum()
```

```python
iota_data= iota_data.dropna()
iota_data.isna().sum()
```

```python
# 데이터 저장
iota_data = iota_data.sort_values('date').reset_index(drop=True)
iota_data.to_csv('iota_train_data')
iota_data
```

```python
flair_iota = iota_data[['date','tweet']]
```

```python
sentimental_tweet = sentimental_analysis(flair_iota)
```

```python
sentimental_tweet.process('flair')
flair_processed_iota = sentimental_tweet.twitter_data.drop('vader',axis=1)
flair_processed_iota
```

```python
flair_processed_iota.to_csv('flair_processed_iota_train_data')
```

```python
pd.read_csv('flair_processed_iota_train_data').drop('Unnamed: 0',axis=1)
```
