## ethereum Train Data - Processed for Flair

- Ï†ÑÏ≤òÎ¶¨ Ïù¥ÌõÑ ÍπÄÎèÑÏòÅ ÌïôÏö∞ÏóêÍ≤å Ï†ÑÎã¨ÌïòÏó¨ ÌïôÏäµÌï† Îç∞Ïù¥ÌÑ∞
- 2022.04.13~2022.04.27

```python
#tweet data 2022.04.14~2022.04.28
date="2022-04-"
d_concat = data_concat()
ethereum_df = []

for i in range(14,29):
    ethereum_df.append(pd.read_csv('./ethereum/ethereum_'+date+str(i), 
                                  encoding='utf-8', on_bad_lines='skip').drop('number',axis=1))
    ethereum_df.append(pd.read_csv('./ethereum/ETH_'+date+str(i), 
                                  encoding='utf-8', on_bad_lines='skip').drop('number',axis=1))

ethereum_data = d_concat.concat(ethereum_df)
```

```python
ethereum_data.isna().sum()
```

```python
ethereum_data= ethereum_data.dropna()
ethereum_data.isna().sum()
```

```python
# Îç∞Ïù¥ÌÑ∞ Ï†ÄÏû•
ethereum_data = ethereum_data.sort_values('date').reset_index(drop=True)
ethereum_data.to_csv('ethereum_train_data')
ethereum_data
```

```python
flair_ethereum = ethereum_data[['date','tweet']]
```

```python
sentimental_tweet = sentimental_analysis(flair_ethereum)
```

```python
sentimental_tweet.process('flair')
flair_processed_ethereum = sentimental_tweet.twitter_data.drop('vader',axis=1)
flair_processed_ethereum
```

```python
flair_processed_ethereum.to_csv('flair_processed_ethereum_train_data')
```

```python
pd.read_csv('flair_processed_ethereum_train_data',lineterminator='\n').drop('Unnamed: 0',axis=1)
```

## 2.1 ethereum Test Data - Processed for Flair

- 2022.04.30~2022.05.14

```python
#tweet data 2022.04.14~2022.04.28
date="2022-05-"
ethereum_test_df = []
d_concat = data_concat()

for i in range(1,16):
    day =''
    if i < 10 : 
        day='0'+str(i)
    else : 
        day = str(i)
    ethereum_test_df.append(pd.read_csv('./ethereum/ethereum_'+date+day, 
                                  encoding='utf-8', on_bad_lines='skip').drop('number',axis=1))
    ethereum_test_df.append(pd.read_csv('./ethereum/ETH_'+date+day, 
                                  encoding='utf-8', on_bad_lines='skip').drop('number',axis=1))

ethereum_test_df.append(pd.read_csv('./ethereum/ethereum_2022-04-30', 
                                  encoding='utf-8', on_bad_lines='skip').drop('number',axis=1))
ethereum_test_df.append(pd.read_csv('./ethereum/ETH_2022-04-30', 
                                  encoding='utf-8', on_bad_lines='skip').drop('number',axis=1))
                                  
ethereum_test_data = d_concat.concat(ethereum_test_df)
ethereum_test_data
```
