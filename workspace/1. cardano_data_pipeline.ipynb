```python
#tweet data 2022.04.14~2022.04.28
date="2022-04-"
d_concat = data_concat()
cardano_df = []

for i in range(14,29):
    cardano_df.append(pd.read_csv('./cardano/ada OR cardano_'+date+str(i), 
                                  encoding='utf-8', on_bad_lines='skip').drop('number',axis=1))

cardano_data = d_concat.concat(cardano_df)
```

```python
sentimental_tweet = sentimental_analysis(flair_cardano)
```

```python
sentimental_tweet.process('flair')
flair_processed_cardano = sentimental_tweet.twitter_data.drop('vader',axis=1)
flair_processed_cardano
```
