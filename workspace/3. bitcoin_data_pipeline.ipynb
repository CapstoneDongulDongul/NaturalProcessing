## 2.1 Bitcoin Train Data - Processed for Flair

- 전처리 이후 김도영 학우에게 전달하여 학습할 데이터
- 2022.04.13~2022.04.27

```python
#tweet data 2022.04.14~2022.04.28
date="2022-04-"
d_concat = data_concat()
bitcoin_df = []

for i in range(14,29):
    bitcoin_df.append(pd.read_csv('./bitcoin/bitcoin_'+date+str(i), 
                                  encoding='utf-8', on_bad_lines='skip').drop('number',axis=1))
    bitcoin_df.append(pd.read_csv('./bitcoin/btc_'+date+str(i), 
                                  encoding='utf-8', on_bad_lines='skip').drop('number',axis=1))

bitcoin_data = d_concat.concat(bitcoin_df)
```

```python
bitcoin_data.isna().sum()
```

```python
bitcoin_data= bitcoin_data.dropna()
bitcoin_data.isna().sum()
```

```python
# 데이터 저장
bitcoin_data = bitcoin_data.sort_values('date').reset_index(drop=True)
bitcoin_data.to_csv('bitcoin_train_data')
bitcoin_data
```

```python
flair_bitcoin = bitcoin_data[['date','tweet']]
```

```python
sentimental_tweet = sentimental_analysis(flair_bitcoin)
```

```python
sentimental_tweet.process('flair')
flair_processed_bitcoin = sentimental_tweet.twitter_data.drop('flair',axis=1)
flair_processed_bitcoin
```

```python
flair_processed_bitcoin.to_csv('flair_processed_bitcoin_train_data')
```

```python
pd.read_csv('flair_processed_bitcoin_train_data').drop('Unnamed: 0',axis=1)
```

## 2.1 Bitcoin Test Data - Processed for Flair

- 2022.04.30~2022.05.14

```python
#tweet data 2022.04.14~2022.04.28
date="2022-05-"
bitcoin_test_df = []
d_concat = data_concat()

for i in range(1,16):
    day =''
    if i < 10 : 
        day='0'+str(i)
    else : 
        day = str(i)
    bitcoin_test_df.append(pd.read_csv('./bitcoin/bitcoin_'+date+day, 
                                  encoding='utf-8', on_bad_lines='skip').drop('number',axis=1))
    bitcoin_test_df.append(pd.read_csv('./bitcoin/btc_'+date+day, 
                                  encoding='utf-8', on_bad_lines='skip').drop('number',axis=1))
    
bitcoin_test_df.append(pd.read_csv('./bitcoin/bitcoin_2022-04-30', 
                                  encoding='utf-8', on_bad_lines='skip').drop('number',axis=1))
bitcoin_test_df.append(pd.read_csv('./bitcoin/btc_2022-04-30', 
                                  encoding='utf-8', on_bad_lines='skip').drop('number',axis=1))
bitcoin_test_data = d_concat.concat(bitcoin_test_df)
bitcoin_test_data
```

```python
bitcoin_test_data.isna().sum()
```
