## 2.1 WAVES Train Data - Processed for Flair

- 전처리 이후 김도영 학우에게 전달하여 학습할 데이터
- 2022.04.13~2022.04.27

```python
#tweet data 2022.04.14~2022.04.28
date="2022-04-"
d_concat = data_concat()
waves_df = []

for i in range(14,29):
    waves_df.append(pd.read_csv('./waves/waves_'+date+str(i), 
                                  encoding='utf-8', on_bad_lines='skip').drop('number',axis=1))
    waves_df.append(pd.read_csv('./waves/waves coin_'+date+str(i), 
                                  encoding='utf-8', on_bad_lines='skip').drop('number',axis=1))

waves_data = d_concat.concat(waves_df)
```

```python
waves_data.isna().sum()
```

```python
waves_data= waves_data.dropna()
waves_data.isna().sum()
```

```python
# 데이터 저장
waves_data = waves_data.sort_values('date').reset_index(drop=True)
waves_data.to_csv('waves_train_data')
waves_data
```

```python
flair_waves = waves_data[['date','tweet']]
```

```python
sentimental_tweet = sentimental_analysis(flair_waves)
```

```python
sentimental_tweet.process('flair')
flair_processed_waves = sentimental_tweet.twitter_data.drop('vader',axis=1)
flair_processed_waves
```

```python
flair_processed_waves.to_csv('flair_processed_waves_train_data')
```

```python
pd.read_csv('flair_processed_waves_train_data').drop('Unnamed: 0',axis=1)
```

## 2.1 WAVES Test Data - Processed for Flair

- 2022.04.30~2022.05.14

```python
#tweet data 2022.04.14~2022.04.28
date="2022-05-"
waves_test_df = []
d_concat = data_concat()

for i in range(1,16):
    day =''
    if i < 10 : 
        day='0'+str(i)
    else : 
        day = str(i)
    waves_test_df.append(pd.read_csv('./waves/waves_'+date+day, 
                                  encoding='utf-8', on_bad_lines='skip').drop('number',axis=1))
    waves_test_df.append(pd.read_csv('./waves/waves coin_'+date+day, 
                                  encoding='utf-8', on_bad_lines='skip').drop('number',axis=1))
waves_test_df.append(pd.read_csv('./waves/waves_2022-04-30', 
                                  encoding='utf-8', on_bad_lines='skip').drop('number',axis=1))
waves_test_df.append(pd.read_csv('./waves/waves coin_'+date+day, 
                                  encoding='utf-8', on_bad_lines='skip').drop('number',axis=1))
waves_test_data = d_concat.concat(waves_test_df)
waves_test_data
```

```python
waves_test_data.isna().sum()
```

```python
waves_test_data = waves_test_data.dropna()
waves_test_data.isna().sum()
```

```python
# 데이터 저장
waves_test_data = waves_test_data.sort_values('date').reset_index(drop=True)
waves_test_data.to_csv('waves_test_data')
waves_test_data
```
