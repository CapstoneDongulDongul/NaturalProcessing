## 2.1 Dogecoin Train Data - Processed for Flair

- 전처리 이후 김도영 학우에게 전달하여 학습할 데이터
- 2022.04.13~2022.04.27

```python
#tweet data 2022.04.14~2022.04.28
date="2022-04-"
d_concat = data_concat()
dogecoin_df = []

for i in range(14,29):
    dogecoin_df.append(pd.read_csv('./dogecoin/doge OR Dogecoin_'+date+str(i), 
                                  encoding='utf-8', on_bad_lines='skip').drop('number',axis=1))

dogecoin_data = d_concat.concat(dogecoin_df)
```

```python
dogecoin_data.isna().sum()
```

```python
dogecoin_data= dogecoin_data.dropna()
dogecoin_data.isna().sum()
```

```python
# 데이터 저장
dogecoin_data = dogecoin_data.sort_values('date').reset_index(drop=True)
dogecoin_data.to_csv('dogecoin_train_data')
dogecoin_data
```

```python
flair_dogecoin = dogecoin_data[['date','tweet']]
```

```python
sentimental_tweet = sentimental_analysis(flair_dogecoin)
```

```python
sentimental_tweet.process('flair')
flair_processed_dogecoin = sentimental_tweet.twitter_data.drop('vader',axis=1)
flair_processed_dogecoin
```

```python
flair_processed_dogecoin.to_csv('flair_processed_dogecoin_train_data')
```

```python
pd.read_csv('flair_processed_dogecoin_train_data').drop('Unnamed: 0',axis=1)
```

## 2.1 ADA(dogecoin) Test Data - Processed for Flair

- 2022.04.30~2022.05.14

```python
#tweet data 2022.04.14~2022.04.28
date="2022-05-"
dogecoin_test_df = []
d_concat = data_concat()

for i in range(1,16):
    day =''
    if i < 10 : 
        day='0'+str(i)
    else : 
        day = str(i)
    dogecoin_test_df.append(pd.read_csv('./dogecoin/doge OR Dogecoin_'+date+day, 
                                  encoding='utf-8', on_bad_lines='skip').drop('number',axis=1))
dogecoin_test_df.append(pd.read_csv('./dogecoin/doge OR Dogecoin_2022-04-30', 
                                  encoding='utf-8', on_bad_lines='skip').drop('number',axis=1))
dogecoin_test_df.append(pd.read_csv('./dogecoin/Dogecoin_2022-05-15', 
                                  encoding='utf-8', on_bad_lines='skip').drop('number',axis=1))
dogecoin_test_data = d_concat.concat(dogecoin_test_df)
dogecoin_test_data
```
