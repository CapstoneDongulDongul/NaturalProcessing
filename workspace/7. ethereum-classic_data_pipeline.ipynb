## 2.1 ethereum-classic Train Data - Processed for Flair

- 전처리 이후 김도영 학우에게 전달하여 학습할 데이터
- 2022.04.13~2022.04.27

```python
#tweet data 2022.04.14~2022.04.28
date="2022-04-"
d_concat = data_concat()
etc_df = []

for i in range(14,29):
    etc_df.append(pd.read_csv('./ethereum-classic/ETC OR ethereum classic_'+date+str(i), 
                                  encoding='utf-8', on_bad_lines='skip').drop('number',axis=1))
    
etc_data = d_concat.concat(etc_df)
```

```python
etc_data.isna().sum()
```

```python
etc_data= etc_data.dropna()
etc_data.isna().sum()
```

```python
# 데이터 저장
etc_data = etc_data.sort_values('date').reset_index(drop=True)
etc_data.to_csv('etc_train_data')
etc_data
```

```python
flair_etc = etc_data[['date','tweet']]
```

```python
sentimental_tweet = sentimental_analysis(flair_etc)
```

```python
sentimental_tweet.process('flair')
flair_processed_etc = sentimental_tweet.twitter_data.drop('vader',axis=1)
flair_processed_etc
```

```python
flair_processed_etc.to_csv('flair_processed_etc_train_data')
```

```python
pd.read_csv('flair_processed_etc_train_data').drop('Unnamed: 0',axis=1)
```
