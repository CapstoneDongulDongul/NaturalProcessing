## 2.1 IOST Train Data - Processed for Flair

- 전처리 이후 김도영 학우에게 전달하여 학습할 데이터
- 2022.04.13~2022.04.27

```python
#tweet data 2022.04.14~2022.04.28
date="2022-04-"
d_concat = data_concat()
iostoken_df = []

for i in range(14,29):
    iostoken_df.append(pd.read_csv('./iostoken/iost_'+date+str(i), 
                                  encoding='utf-8', on_bad_lines='skip').drop('number',axis=1))

iostoken_data = d_concat.concat(iostoken_df)
```

```python
iostoken_data.isna().sum()
```

```python
iostoken_data= iostoken_data.dropna()
iostoken_data.isna().sum()
```

```python
# 데이터 저장
iostoken_data = iostoken_data.sort_values('date').reset_index(drop=True)
iostoken_data.to_csv('iostoken_train_data')
iostoken_data
```

```python
flair_iostoken = iostoken_data[['date','tweet']]
```

```python
sentimental_tweet = sentimental_analysis(flair_iostoken)
```

```python
sentimental_tweet.process('flair')
flair_processed_iostoken = sentimental_tweet.twitter_data.drop('vader',axis=1)
flair_processed_iostoken
```

```python
flair_processed_iostoken.to_csv('flair_processed_iostoken_train_data')
```

```python
pd.read_csv('flair_processed_iostoken_train_data').drop('Unnamed: 0',axis=1)
```

## 2.1 iostoken Test Data - Processed for Flair
