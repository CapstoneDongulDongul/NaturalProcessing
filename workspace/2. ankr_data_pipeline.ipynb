## 2.1 ANKR Train Data - Processed for Flair

- 전처리 이후 김도영 학우에게 전달하여 학습할 데이터
- 2022.04.13~2022.04.27

```python
#tweet data 2022.04.14~2022.04.28
date="2022-04-"
d_concat = data_concat()
ankr_df = []

for i in range(14,29):
    ankr_df.append(pd.read_csv('./ankr/ankr_'+date+str(i), 
                                  encoding='utf-8', on_bad_lines='skip').drop('number',axis=1))

ankr_data = d_concat.concat(ankr_df)
```

```python
ankr_data.isna().sum()
```

```python
ankr_data= ankr_data.dropna()
ankr_data.isna().sum()
```

```python
# 데이터 저장
ankr_data = ankr_data.sort_values('date').reset_index(drop=True)
ankr_data.to_csv('ankr_train_data')
ankr_data
```

```python
flair_ankr = ankr_data[['date','tweet']]
```

```python
sentimental_tweet = sentimental_analysis(flair_ankr)
```

```python
sentimental_tweet.process('flair')
flair_processed_ankr = sentimental_tweet.twitter_data.drop('vader',axis=1)
flair_processed_ankr
```

```python
flair_processed_ankr.to_csv('flair_processed_ankr_train_data')
```

```python
pd.read_csv('flair_processed_ankr_train_data').drop('Unnamed: 0',axis=1)
```

## 2.1 ANKR Test Data - Processed for Flair

- 2022.04.30~2022.05.14

```python
#tweet data 2022.04.14~2022.04.28
date="2022-05-"
ankr_test_df = []
d_concat = data_concat()

for i in range(1,16):
    day =''
    if i < 10 : 
        day='0'+str(i)
    else : 
        day = str(i)
    ankr_test_df.append(pd.read_csv('./ankr/ankr_'+date+day, 
                                  encoding='utf-8', on_bad_lines='skip').drop('number',axis=1))
ankr_test_df.append(pd.read_csv('./ankr/ankr_2022-04-30', 
                                  encoding='utf-8', on_bad_lines='skip').drop('number',axis=1))
ankr_test_data = d_concat.concat(ankr_test_df)
ankr_test_data
```
