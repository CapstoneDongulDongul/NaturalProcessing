## 2.1 SANDBOX Train Data - Processed for Flair

- Ï†ÑÏ≤òÎ¶¨ Ïù¥ÌõÑ ÍπÄÎèÑÏòÅ ÌïôÏö∞ÏóêÍ≤å Ï†ÑÎã¨ÌïòÏó¨ ÌïôÏäµÌï† Îç∞Ïù¥ÌÑ∞
- 2022.04.13~2022.04.27

```python
#tweet data 2022.04.14~2022.04.28
date="2022-04-"
d_concat = data_concat()
sandbox_df = []

for i in range(14,29):
    sandbox_df.append(pd.read_csv('./the-sandbox/sandbox coin OR sand coin_'+date+str(i), 
                                  encoding='utf-8', on_bad_lines='skip').drop('number',axis=1))
    sandbox_df.append(pd.read_csv('./the-sandbox/sandbox OR SAND_'+date+str(i), 
                                  encoding='utf-8', on_bad_lines='skip').drop('number',axis=1))
sandbox_data = d_concat.concat(sandbox_df)
```

```python
sandbox_data.isna().sum()
```

```python
sandbox_data= sandbox_data.dropna()
sandbox_data.isna().sum()
```

```python
# Îç∞Ïù¥ÌÑ∞ Ï†ÄÏû•
sandbox_data = sandbox_data.sort_values('date').reset_index(drop=True)
sandbox_data.to_csv('sandbox_train_data')
sandbox_data
```

```python
flair_sandbox = sandbox_data[['date','tweet']]
```

```python
sentimental_tweet = sentimental_analysis(flair_sandbox)
```

```python
sentimental_tweet.process('flair')
flair_processed_sandbox = sentimental_tweet.twitter_data.drop('vader',axis=1)
flair_processed_sandbox
```

```python
sentimental_tweet.sentimental_with_flair()
```

```python
data = sentimental_tweet.twitter_data.drop('vader',axis=1)
data.to_csv('flair_sandbox_train_data.csv')
```

```python
flair_processed_sandbox.to_csv('flair_processed_sandbox_train_data')
```

## 2.1 SANDBOX Test Data - Processed for Flair

- 2022.04.30~2022.05.14

```python
#tweet data 2022.04.14~2022.04.28
date="2022-05-"
sandbox_test_df = []
d_concat = data_concat()

for i in range(1,16):
    day =''
    if i < 10 : 
        day='0'+str(i)
    else : 
        day = str(i)
    sandbox_test_df.append(pd.read_csv('./the-sandbox/sandbox coin OR sand coin_'+date+day, 
                                  encoding='utf-8', on_bad_lines='skip').drop('number',axis=1))
sandbox_test_df.append(pd.read_csv('./the-sandbox/sandbox OR SAND_2022-04-30', 
                                  encoding='utf-8', on_bad_lines='skip').drop('number',axis=1))
sandbox_test_data = d_concat.concat(sandbox_test_df)
sandbox_test_data
```
